```python
Given an integer number n, return the difference between the product of its digits and the sum of its digits.
 

Example 1:

Input: n = 234
Output: 15 
Explanation: 
Product of digits = 2 * 3 * 4 = 24 
Sum of digits = 2 + 3 + 4 = 9 
Result = 24 - 9 = 15
Example 2:

Input: n = 4421
Output: 21
Explanation: 
Product of digits = 4 * 4 * 2 * 1 = 32 
Sum of digits = 4 + 4 + 2 + 1 = 11 
Result = 32 - 11 = 21
 

Constraints:

1 <= n <= 10^5

Success
Details 
Runtime: 50 ms, faster than 36.71% of Python3 online submissions for Subtract the Product and Sum of Digits of an Integer.
Memory Usage: 13.9 MB, less than 52.15% of Python3 online submissions for Subtract the Product and Sum of Digits of an Integer.
```

```python
def subtractProductAndSum( n: int) -> int:

    strnum = str(n)
    sum = 0
    product = 1

    for i in range(0, len(strnum)):
      sum = sum + (ord(strnum[i])-48)
      product = product * (ord(strnum[i])-48)

    return( product - sum )

print(subtractProductAndSum(4421))
```
