```python
# create a matrix of zeroes  5x5 in size
gradient = np.zeros((5,5))

# fill the diagonal with 1s
for i in range(0,5):
  gradient[i,i] = 1

print(gradient)
plt.imshow(gradient)
```

```python
# create a matrix of zeroes  15x15 in size
gradient = np.zeros((15,15))


for i in range(0,15):
  for j in range(0,15):
    gradient[i,j] = i + j

plt.imshow(gradient)
```

```python
# create a matrix of zeroes  15x15 in size
gradient = np.zeros((15,15))

for i in range(0,15):
  for j in range(0,15):
    gradient[i,j] = abs(i-7) + abs(j-7)

plt.imshow(gradient)
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace( (-1)*np.pi,  np.pi , 50 )
ysin = np.sin(x)
ycos = np.cos(x)
plt.plot(x,ysin,"r")   # plot y(x) with red marker
plt.plot(x,ycos,"b")   # plot y(x) with blue marker
plt.plot(x,ycos*ycos,"g")   # plot y(x) with green marker
plt.plot(x,ysin*ysin,"y")   # plot y(x) with yellow marker

plt.plot(x,(-1)*ycos*ycos,"m")   # plot -y(x) with magenta marker
plt.plot(x,(-1)*ysin*ysin,"c")   # plot -y(x) with cyan marker
```
